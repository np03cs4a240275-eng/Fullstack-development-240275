```python
import numpy as np

A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)
AA_inv = np.matmul(A, A_inv)
print("A * A^{-1} =\n", AA_inv)

AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nAB =\n", AB)
print("\nBA =\n", BA)
print("\nIs AB equal to BA? ->", np.allclose(AB, BA))

AB_T = AB.T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T =\n", AB_T)
print("\nB^T A^T =\n", BT_AT)
print("\nIs (AB)^T equal to B^T A^T? ->", np.allclose(AB_T, BT_AT))

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])
b_sys = np.array([-1, -3, 9])

solution_solve = np.linalg.solve(A_sys, b_sys)
A_sys_inv = np.linalg.inv(A_sys)
solution_inv = np.dot(A_sys_inv, b_sys)

print("\nSolution using np.linalg.solve:\n", solution_solve)
print("\nSolution using inverse A^{-1}B:\n", solution_inv)
print("\nAre both solutions equal? ->", np.allclose(solution_solve, solution_inv))
```

```python
import time

# Helper sizes
N = 1_000_000
M = 1000

# 4.2.1. Element-wise Addition
# Python lists
list_a = list(range(N))
list_b = list(range(N))

start = time.time()
list_add = [list_a[i] + list_b[i] for i in range(N)]
end = time.time()
time_list_add = end - start

# NumPy arrays
arr_a = np.arange(N)
arr_b = np.arange(N)

start = time.time()
arr_add = arr_a + arr_b
end = time.time()
time_numpy_add = end - start

print("Element-wise addition - Python lists time:", time_list_add)
print("Element-wise addition - NumPy arrays time:", time_numpy_add)

# 4.2.2. Element-wise Multiplication
# Python lists
start = time.time()
list_mul = [list_a[i] * list_b[i] for i in range(N)]
end = time.time()
time_list_mul = end - start

# NumPy arrays
start = time.time()
arr_mul = arr_a * arr_b
end = time.time()
time_numpy_mul = end - start

print("Element-wise multiplication - Python lists time:", time_list_mul)
print("Element-wise multiplication - NumPy arrays time:", time_numpy_mul)

# 4.2.3. Dot Product
# Python lists
start = time.time()
dot_list = sum(list_a[i] * list_b[i] for i in range(N))
end = time.time()
time_list_dot = end - start

# NumPy arrays
start = time.time()
dot_numpy = np.dot(arr_a, arr_b)
end = time.time()
time_numpy_dot = end - start

print("Dot product - Python lists time:", time_list_dot)
print("Dot product - NumPy arrays time:", time_numpy_dot)

# 4.2.4. Matrix Multiplication
# Python lists
matrix_A_list = [[1.0] * M for _ in range(M)]
matrix_B_list = [[1.0] * M for _ in range(M)]

start = time.time()
matrix_C_list = [[sum(matrix_A_list[i][k] * matrix_B_list[k][j] for k in range(M))
                  for j in range(M)]
                 for i in range(M)]
end = time.time()
time_list_matmul = end - start

# NumPy arrays
matrix_A_np = np.ones((M, M))
matrix_B_np = np.ones((M, M))

start = time.time()
matrix_C_np = np.dot(matrix_A_np, matrix_B_np)
end = time.time()
time_numpy_matmul = end - start

print("Matrix multiplication (1000x1000) - Python lists time:", time_list_matmul)
print("Matrix multiplication (1000x1000) - NumPy arrays time:", time_numpy_matmul)
```
